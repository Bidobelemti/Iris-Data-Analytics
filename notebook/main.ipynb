# Dataset Iris

```python
import pandas as pd
import sys
import os

sys.path.append(os.path.abspath(".."))

path = '../data/Iris.csv'
df_raw = pd.read_csv(path, index_col=0) # importamos el dataset y colocamos a la columna ID como id del dataset
print('Número de elementos: ',len(df_raw))
print('Número de caracteristicas: ', len(df_raw.columns))
print('--------------------------------------------------')
print(df_raw.info())
print('--------------------------------------------------')
print('Existen nulos')
print(df_raw.isnull().sum())
print('--------------------------------------------------')
print('Existen duplicados')
print(df_raw.duplicated().sum())
df = df_raw.drop_duplicates()
print(df['Species'].unique())
print('Será nuestro valor a predecir, en base a 4 parametros, longitud y ancho de sus pétalos y sépalos.')
```

```python
from src.visualization import show_hist, show_bloxplots, show_relation_length_width
show_hist(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,8))
sns.scatterplot(
    data=df,
    x='SepalLengthCm',
    y='SepalWidthCm',
    hue='Species',
    palette='deep'
)
plt.title("Relación entre largo y ancho del sépalo")
plt.show()
```
